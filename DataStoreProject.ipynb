<font size = 5px><b>Pie Chart </b></font>

```python
#pie chart for Sales Distribution by Segment
sales_by_segment = data.groupby('Segment')['Sales'].sum()
sales_by_segment.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 4))
plt.title('Sales Distribution by Segment')

#setting legend properties
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))
plt.show()

#pie chart for Sales Distribution by Category
sales_by_segment = data.groupby('Category')['Sales'].sum()
sales_by_segment.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 4))
plt.title('Sales Distribution by Category')

#setting legend properties
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))
plt.show()

```
